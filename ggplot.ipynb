## Principe de base

## Exemples

* Exemple 1: 
    1. Plotter `carat` en fonction de `price` en utilisant des points
    2. Colorier les point en fonction de `cut`
    3. La taille des points devrait proportionnelle à `table`


```python
p = ggplot(aes(x="carat", y="price", color="cut"), data=diamonds)
```

```python
x = p + geom_point()
```

* Exemple 2:
    1. Plotter le kernel density de `price` en coloriant par `cut` (avec fill)
    2. Utiliser des facets en fonction de `clarity` (facet_wrap ou facet_grid)

```python
diamonds.clarity.unique()
```

```python
ggplot(aes(x="price", color="cut"), data=diamonds) + geom_density() + facet_wrap("clarity")
```

* Exemple 3:

    1. représenter la puissance des voitures (hp) en fonction de leur (poids)
    1. avec un scatter plot,
    1. superposer les noms des voitures (label),
    1. mettre un titre,
    1. tracer une ligne horizontale à la valeur moyenne de la puissance (yintercept)
    1. renommer les labels des axes x et y avec les bonnes unités (xlab et ylab)

```python
ggplot(aes(x="wt", y="hp", label="name"), data=mtcars) +\
    geom_point(color="steelblue") + geom_text(alpha=0.5) + ggtitle("Puissance des voitures") +\
    geom_hline(yintercept=mtcars.hp.mean(), color="red")
```
