# La visualisation des données pour tous en Python

La première librairie de visualisation de données qui vient à l'esprit de presque tout développeur Python est `matplotlib`. 
En effet elle a longtemps été la plus pratique et la plus complète mais il existe aujourd'hui des alternatives qui peuvent s'avérer un meilleur choix en fonction d'un besoin particulier.

Tout au long de cette présentation nous nous intéresserons à 3 librairies très utilisées par des chercheurs dans différentes disciplines.

Le but n'est pas de remplacer `matplotlib` mais d'apporter des solutions à des problématiques spécifiques (ex: l'analyse statistiques) voire même de nouvelles solutions (ex: façon différente de penser la visualisation, interactivité...)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Les données

```python
mtcars = pd.read_csv("data/mtcars.csv")
mtcars.head()
```

Voici un exemple de code utilisant `matplotlib` pour afficher deux plots avec ce dataset:

## Le code

```python
fig = plt.figure(figsize=(16, 8))

ax1 = fig.add_subplot(121)
ax1.scatter(mtcars.wt, mtcars.mpg, color="steelblue")
ax1.set_title('Distance parcourue par les voitures')
ax1.set_xlabel('Poids (livres)')
ax1.set_ylabel('Distance (miles/gallon)');

ax2 = fig.add_subplot(122)
ax2.scatter(mtcars.wt, mtcars.hp, color="indianred")
ax2.set_title('Puissance des voitures')
ax2.set_xlabel('Poids (livres)')
ax2.set_ylabel('Puissance (chevaux)');

fig.suptitle('Quelques parametres des voitures issues du magazine Motor Trend de 1974');
```

---

Matplotlib: 

* est très générale

* est une librairie "de bas niveau"

* ne fournit pas d'interactivité avec les données

3 Alternatives:

* **seaborn**: une librairie pour l'analyse statistique
* **ggplot**: permet de créer des plots en utilisant une grammaire
* **bokeh**: librairie moderne qui permet d'interagir avec les données
